```python
#function to create the bering strait transect
def create_transect(starty, startx, length, var,mask):
    CanSib = np.zeros((50,length)) #2D matrix of horizonal vs depth
    CanSib_mask = np.zeros((50,length))
    label = np.zeros((length))
    Blons = np.zeros((length))
    Blats = np.zeros((length))

    for i in range(0,length):
        CanSib[:,i] = var[:,starty-i,startx+i] #starts at the last index for lat and first for long
        CanSib_mask[:,i] = mask[:,starty-i,startx+i]
        Blons[i] = lon[starty-i,startx+i]#Lines up lon and lat to have the matching indices
        Blats[i] = lat[starty-i,startx+i]#such that Blon[0]+Blat[0]= start of transect, and as long as
        #the indices match, the coordinates will follow the transect line exactly
      
        #label[i] = lat[starty+i,startx+i]
        
    CanSib = ma.array(CanSib[::],mask=(CanSib_mask-1)*-1)
    print(np.shape(CanSib))
    return CanSib, Blons, Blats
```

```python
#full
Bertranc_kap, Blons, Blats = create_transect(799,100,400,ckappa,mask) #starty,startx,length,var,mask
Bertransj_kap, Blons, Blats = create_transect(799,100,400,sjkappa,mask)
Bertranp_kap, Blons, Blats = create_transect(799,100,400,pkappa,mask)
#tidal
Bertransj_kapt, Blons, Blats = create_transect(799,100,400,sjkaptidal,mask)
Bertranp_kapt, Blons, Blats = create_transect(799,100,400,pkaptidal,mask)
```

```python
#calculate along strack distance->the total distance of the transect
alongtrack = np.zeros(length,dtype=int)
distance = 0 
for index in np.arange(0,length-1):
    start = (Blats[index],Blons[index]) #indices of the locations for the gyre, the ridge, and the basin
    end = (Blats[index+1],Blons[index+1])
    distance += round(geodesic(start, end).kilometers) #syntax like d=d+1
    #round to the nearest 5 to make the labesl cleaner
    alongtrack[index+1] = int(round(distance/5.0)*5.0)
```
